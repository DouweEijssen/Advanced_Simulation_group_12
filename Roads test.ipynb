Plot one road

```python
import matplotlib.pyplot as plt

plt.style.use('default')


def plot_rd(name, df):
    plt.figure()
    plt.plot(df['LON'], df['LAT'])
    plt.title(name)
    plt.show()
```

```python
for rd in dfroads.itertuples():

    # print(rd)
    rd_nr = rd[0]      # index
    rd_name = rd[1]

    data = []

    for i in range(2, len(rd), 3):
        # print(i)

        lrp = (rd[i], rd[i+1], rd[i+2])
        # print(lrp)

        if pd.isna(rd[i]):
            break
        else:
            data.append(lrp)

    # print(data)

    df_rd = pd.DataFrame(data, columns=['LRP', 'LAT', 'LON'])

    # reset lrp_changed_index, which is used later when LRPs are corrected (changed) for roads
    lrp_changed_index = []

    plot_rd(rd_name, df_rd)

    # use break to test only (rd_nr=1) road, remove break to plot all roads
    # break
    if rd_nr == 0:
        break
```
